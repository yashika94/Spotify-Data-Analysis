# Data Normalization

```python
#Checking the Ranges of the predictor variables and dependent variable
plt.figure(figsize=(20,7))
sns.boxplot(data=df)
```

Normalising the data for Year column beacuse the value is too high when compared to other independent variable

```python

# Create x to store scaled values as floats
x = df[['Year']].values.astype(float)

# Preparing for normalizing
min_max_scaler = preprocessing.MinMaxScaler()

# Transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(x)

# Run the normalizer on the dataframe
df[['Year']] = pd.DataFrame(x_scaled)
```

```python
from sklearn.model_selection import  train_test_split

X = df[ ['Beats Per Minute (BPM)', 'Energy', 'Danceability', 'Loudness (dB)', 'Valence',
       'Liveness','Length (Duration)', 'Acousticness','Speechiness']].replace(',', '', regex = True).astype(float)

y = df['Popularity']

#Spliting data into Training set, Validation set and Test set 20%

X_t, X_test, y_t, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

X_train, X_val, y_train, y_val = train_test_split(X_t, y_t, test_size=0.15, random_state=1)
```

Ans- No, the training and test data sets are typically different. The training data set is used to train the model, while the test data set is used to evaluate the performance of the trained model. The test data set is typically a smaller and independent set of data, that is not used during the training process, and is used to give an unbiased estimate of the model's performance on unseen data.

# Making predictions

```python
# Make predictions using the training set
y_pred = regr.predict(X_train)
# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_train, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_train, y_pred))
r2 = r2_score(y_train,y_pred)
print('R^2 score on training set =',r2)
```

```python
# Make predictions using the testing set
y_pred = regr.predict(X_test)
# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_test, y_pred))
r2 = r2_score(y_test,y_pred)
print('R^2 score on test set =',r2)
```

```python
# Make predictions using the validation set
y_pred = regr.predict(X_val)

# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_val, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_val, y_pred))
r2_val = r2_score(y_val,y_pred)
print('R^2 score on validation set =',r2_val)
```

```python
#Imputation using KNN, removed 5% of data from Danceability column 
from sklearn.impute import KNNImputer
from sklearn.impute import SimpleImputer
df_mainImpute = df.copy(deep=True)
def removeRandomData(dataset, percent, col):
    dataset.loc[dataset.sample(frac = percent).index, col] = np.nan


df_mainImpute['Popularity10'] = df_mainImpute['Popularity']
df_mainImpute['Popularity5'] = df_mainImpute['Popularity']
df_mainImpute['Popularity1'] = df_mainImpute['Popularity']

removeRandomData(df_mainImpute, 0.10, 'Popularity10')
removeRandomData(df_mainImpute, 0.05, 'Popularity5')
removeRandomData(df_mainImpute, 0.01, 'Popularity1')
print(df_mainImpute.isnull().sum())
```

# * Remove outliers and keep outliers (does if have an effect of the final predictive model)?

# Linear Regression without outliers

```python
from scipy import stats

def removeOutliers(df):
  data = df[(np.abs(stats.zscore(df)) < 3).all(axis=1)]
  return data

from sklearn.model_selection import  train_test_split
df = df_copy[['Beats Per Minute (BPM)', 'Energy', 'Danceability', 'Loudness (dB)', 'Valence','Liveness','Length (Duration)', 'Acousticness','Speechiness', 'Popularity']].replace(',', '', regex = True).astype(int)
#df['Popularity'] 

df_noOutliers = removeOutliers(df)
X = df_noOutliers[ ['Beats Per Minute (BPM)', 'Energy', 'Danceability', 'Loudness (dB)', 'Valence',
       'Liveness','Length (Duration)', 'Acousticness','Speechiness']].replace(',', '', regex = True).astype(int)

y = df_noOutliers['Popularity']

#Spliting data into Training set, Validation set and Test set 20%

X_t, X_test, y_t, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

X_train, X_val, y_train, y_val = train_test_split(X_t, y_t, test_size=0.15, random_state=1)

  # Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train,y_train)
```

```python
# Make predictions using the training set
y_pred = regr.predict(X_train)
# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_train, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_train, y_pred))
r2 = r2_score(y_train,y_pred)
print('R^2 score on training set =',r2)
```

```python
# Make predictions using the testing set
y_pred = regr.predict(X_test)
# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_test, y_pred))
r2 = r2_score(y_test,y_pred)
print('R^2 score on test set =',r2)
```

```python
# Make predictions using the validation set
y_pred = regr.predict(X_val)

# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_val, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_val, y_pred))
r2_val = r2_score(y_val,y_pred)
print('R^2 score on validation set =',r2_val)
```
